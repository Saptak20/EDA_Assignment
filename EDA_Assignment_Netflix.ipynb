Task 1: Data Overview

```python
import pandas as pd
df = pd.read_csv('netflix_customer_churn.csv')
df.head()
df.isnull().sum()
df.nunique()
{col: df[col].unique().tolist() for col in ["gender", "subscription_type", "region", "device", "payment_method", "favorite_genre"]}
```

Task 2: Univariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")

sns.histplot(df['age'], kde=True); plt.show()
sns.histplot(df['watch_hours'], kde=True); plt.show()
sns.histplot(df['monthly_fee'], kde=True); plt.show()
sns.countplot(x='churned', data=df); plt.show()

sns.countplot(x='subscription_type', data=df); plt.show()
sns.countplot(x='gender', data=df); plt.show()
sns.countplot(x='region', data=df); plt.xticks(rotation=30); plt.show()
sns.countplot(x='device', data=df); plt.show()
sns.countplot(x='payment_method', data=df); plt.xticks(rotation=30); plt.show()
sns.countplot(x='favorite_genre', data=df); plt.xticks(rotation=30); plt.show()
```

Task 3: Bivariate Analysis

```python
sns.barplot(x='subscription_type', y='watch_hours', data=df); plt.show()
sns.barplot(x='subscription_type', y='monthly_fee', data=df); plt.show()
sns.barplot(x='region', y='watch_hours', data=df); plt.xticks(rotation=30); plt.show()
sns.barplot(x='region', y='monthly_fee', data=df); plt.xticks(rotation=30); plt.show()
sns.barplot(x='device', y='watch_hours', data=df); plt.show()
sns.barplot(x='device', y='monthly_fee', data=df); plt.show()
sns.barplot(x='favorite_genre', y='avg_watch_time_per_day', data=df); plt.xticks(rotation=30); plt.show()

sns.barplot(x='gender', y='churned', data=df); plt.show()
sns.barplot(x='region', y='churned', data=df); plt.xticks(rotation=30); plt.show()
sns.barplot(x='subscription_type', y='churned', data=df); plt.show()
sns.barplot(x='payment_method', y='churned', data=df); plt.xticks(rotation=30); plt.show()
```

Task 4: Correlation Analysis

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm'); plt.show()
```

Task 5: Insights

1) Customers with low watch hours tend to churn more.

2) Premium users churn less than Basic users.

3) Africa has the highest average watch hours.

4) TV is the most-used device.

5) Gift Card and Crypto are the most-used payment methods.
